```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('PeopleDataset.csv')
df.head()
```

```python
plt.bar(df['FirstName'], df['Height_cm'])
plt.xlabel('First Name')
plt.ylabel('Height (cm)')
plt.title('Height of Individuals')
plt.xticks(rotation=45)
```

```python
df['FirstName'].count()
df['Age'].mean()
df['Age'].min()
df['Height_cm'].max()
gender_counts = df['Gender'].value_counts()
print(f'Females count = {gender_counts.get("Female", 0)}')
print(f'Males count = {gender_counts.get("Male", 0)}')


```
